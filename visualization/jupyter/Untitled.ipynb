```python

json_file = open('../covidbr-vis/src/assets/json/codufOriginal.json', 'r')
state_data = json.load(json_file)
```

```python
for state in state_data['features']:
    for coordinates in state['geometry']['coordinates']:
        for coordinate in coordinates:
            width = 920
            heigth = 960
            if len(coordinate) > 2:
                for coord in coordinate:
#                     print(len(coord))
                    coord[0] = (coord[0] + 180) * (width / 360)
                    latRad = coord[1] * math.pi / 180
                    mercN = math.log(math.tan((math.pi / 4) + (latRad / 2)))
                    coord[1] = (heigth / 2) - (heigth * mercN / (2 * math.pi))
                #coordinate[len(coordinate)-1] = coordinate[0]
            else:
#                 print(len(coordinate))
                coordinate[0] = (coordinate[0] + 180) * (width / 360)
                latRad = coordinate[1] * math.pi / 180
                mercN = math.log(math.tan((math.pi / 4) + (latRad / 2)))
                coordinate[1] = (heigth / 2) - (heigth * mercN / (2 * math.pi))
        #coordinates[len(coordinates)-1] = coordinates[0]

        
min_x = 10000000
min_y = 10000000
max_x = 0
max_y = 0
for state in state_data['features']:
    for coordinates in state['geometry']['coordinates']:
        for coordinate in coordinates:
            if len(coordinate) > 2:
                for coord in coordinate:
                    min_x = min(coord[0], min_x)
                    min_y = min(coord[1], min_y)
                    max_x = max(coord[0], max_x)
                    max_y = max(coord[1], max_y)
            else:
                min_x = min(coordinate[0], min_x)
                min_y = min(coordinate[1], min_y)
                max_x = max(coordinate[0], max_x)
                max_y = max(coordinate[1], max_y)


scale_ratio = 920/max(max_x-min_x, max_y-min_y)

for mun in state_data['features']:
    for coordinates in mun['geometry']['coordinates']:
        for coordinate in coordinates:
            if len(coordinate) > 2:
                for coord in coordinate:
                    coord[0]-= min_x
                    coord[1]-= min_y
            else:
                coordinate[0]-=min_x
                coordinate[1]-=min_y

for state in state_data['features']:
    for coordinates in state['geometry']['coordinates']:
        for coordinate in coordinates:
            if len(coordinate) > 2:
                for coord in coordinate:
                    coord[0]*= scale_ratio
                    coord[1]*= scale_ratio
            else:
                coordinate[0]*=scale_ratio
                coordinate[1]*=scale_ratio
```
